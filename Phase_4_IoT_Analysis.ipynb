```python
#Importa librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga dataset
df = pd.read_csv('iot_telemetry_data.csv')
```

```python
# Ver primeras Lineas
df.head()
```

```python
# Revisa la información general
df.info()
```

```python
#Revisar datos faltantes
df.isnull().sum()
```

```python
#Eliminar valores nulos
df = df.dropna()
```

```python
#Estadísticas descriptivas
df.describe()
```

```python
#Crea muestra del dataset
sample_df = df.sample(1000)
```

```python
# Grafica de temperatura
plt.figure(figsize=(10,5))
plt.plot(sample_df['temp'])
plt.title('Temperature Behavior')
plt.xlabel('Sample Records')
plt.ylabel('Temperature')
plt.show()
```

```python
# Gráfica de Humedad
plt.figure(figsize=(10,5))
plt.plot(sample_df['humidity'])
plt.title('Humidity Behavior')
plt.xlabel('Sample Records')
plt.ylabel('Humidity')
plt.show()
```

```python
# Temperatura Vs Humendad
plt.figure(figsize=(8,6))
sns.scatterplot(x=sample_df['temp'], y=sample_df['humidity'])
plt.title('Temperature vs Humidity')
plt.show()
```

```python
# Heatmap de correlación
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Gráfica de humo
plt.figure(figsize=(10,5))
plt.plot(sample_df['smoke'])
plt.title('Smoke Levels')
plt.xlabel('Sample Records')
plt.ylabel('Smoke')
plt.show()
```

```python
# Gráfica monoxido de carbono
plt.figure(figsize=(10,5))
plt.plot(sample_df['co'])
plt.title('Carbon Monoxide Levels')
plt.xlabel('Sample Records')
plt.ylabel('CO')
plt.show()
```
